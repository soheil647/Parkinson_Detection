```python
def select_n_components(var_ratio, goal_var: float) -> int:
    # Set initial variance explained so far
    total_variance = 0.0
    
    # Set initial number of features
    n_components = 0
    
    # For the explained variance of each feature:
    for explained_variance in var_ratio:
        
        # Add the explained variance to the total
        total_variance += explained_variance
        
        # Add one to the number of components
        n_components += 1
        
        # If we reach our goal level of explained variance
        if total_variance >= goal_var:
            # End the loop
            break
            
    # Return the number of components
    return n_components
```

```python
# PCA Function
def pca_func(in_data, components = 0.95, whitening = False, svd_solver = 'full', plot = True):
  #draw a plot
  if plot:
    pca_base = PCA().fit(in_data)
    plt.plot(pca_base.explained_variance_ratio_)
    plt.xlabel('n_components')
    plt.ylabel('Variance')
    plt.show()
  #main PCA
  main_pca = PCA(n_components = components, svd_solver = svd_solver, whiten = whitening).fit(in_data)
  out_data = main_pca.transform(in_data)
  out_data = pd.DataFrame(out_data)
  return out_data

# LDA Function
def lda_func(in_data, in_target, components = None, solver = 'svd', plot = True):
  # if plot:
  #   lda_base = LinearDiscriminantAnalysis(n_components = components).fit(in_data, in_target)
  #   var_ratio = lda_base.explained_variance_ratio_
  #   plt.plot(var_ratio)
  #   plt.xlabel('n_components')
  #   plt.ylabel('Variance')
  #   plt.show()
  # components = select_n_components(var_ratio, 0.95)
  main_lda = LinearDiscriminantAnalysis(n_components = components, solver = solver).fit(in_data, in_target)
  out_data = main_lda.transform(in_data)
  out_data = pd.DataFrame(out_data)
  return out_data
  
# ICA Function
def ica_func(in_data, components = 78, whitening = False, random = 0):
  main_ica = FastICA(n_components = components, random_state=random).fit(in_data)
  out_data = main_ica.transform(in_data)
  out_data = pd.DataFrame(out_data)
  return out_data
```

```python
# Oversampling
def oversample(in_data, target, sampling_strategy='minority'):
  oversample = RandomOverSampler(sampling_strategy= sampling_strategy)
  in_data_over, target_over = oversample.fit_resample(in_data, target)
  in_data_over = pd.DataFrame(in_data_over)
  target_over = pd.DataFrame(target_over)
  return in_data_over, target_over

# Undersampling
def undersample(in_data, target, sampling_strategy= 'majority'):
  undersample = RandomUnderSampler(sampling_strategy= sampling_strategy)
  in_data_under, target_under = undersample.fit_resample(in_data, target)
  in_data_under = pd.DataFrame(in_data_under)
  target_under = pd.DataFrame(target_under)
  return in_data_under, target_under

# SMOTE
def smote(in_data, target, sampling_strategy= 'minority'):
  smote = SMOTE(sampling_strategy= sampling_strategy)
  in_data_smote, target_smote = smote.fit_resample(in_data, target)
  in_data_smote = pd.DataFrame(in_data_smote)
  target_smote = pd.DataFrame(target_smote)
  return in_data_smote, target_smote
```

```python
balance_data, balance_target = smote(df_normalized, target)
display(balance_data)
display(balance_target)
```

```python
# GMM Model
def gmm_classifier(in_data, components = 2, random_state = 0, covariance_type = 'full'):
  gmm = GaussianMixture(n_components=components, random_state= random_state, covariance_type = covariance_type).fit(in_data)
  model_prediction = gmm.predict(in_data)
  model_prediction = pd.DataFrame(model_prediction)
  return model_prediction
```
